## Handling missing values

```python
df['rectal_temp'] = df['rectal_temp'].fillna(df['rectal_temp'].median())
df['pulse'] = df['pulse'].fillna(df['pulse'].median())
df['respiratory_rate'] = df['respiratory_rate'].fillna(df['respiratory_rate'].median())
```

```python
#Categorical values so we use mode to fill missing values
df['pain']=df['pain'].fillna(df['pain'].mode()[0])
df['mucous_membrane']=df['mucous_membrane'].fillna(df['mucous_membrane'].mode()[0])
df['capillary_refill_time']=df['capillary_refill_time'].fillna(df['capillary_refill_time'].mode()[0])
```

```python
# this column has more number of missing values so we are filling it with Unknown
df['nasogastric_tube']=df['nasogastric_tube'].fillna('Unknown')
df['abdomen']=df['abdomen'].fillna('Unknown')
```

```python
category = ['temp_of_extremities', 'peripheral_pulse', 'peristalsis', 'abdominal_distention',
            'nasogastric_tube', 'nasogastric_reflux', 'rectal_exam_feces', 'abdomen', 'abdomo_appearance']
for i in category:
    df[i] = df[i].fillna(df[i].mode()[0])
```

```python
column = ['packed_cell_volume', 'total_protein']
for i in column:
    df[i] = df[i].fillna(df[i].median())
```

Dropping columns with more number of missing values

```python
df.drop(columns=['abdomo_appearance'], inplace=True)
```

```python
df.drop(columns=['abdomo_protein'], inplace=True)
```

```python
df.drop(columns=['nasogastric_reflux_ph'], inplace=True)
```
